```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('Creditcard_data.csv')
df.head(10)
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
df.iloc[:, :-1] = scaler.fit_transform(df.iloc[:, :-1])
df.head(10)
```

```python
from imblearn.over_sampling import SMOTE, ADASYN


# separate the features and target variable
X = df.drop('Class', axis=1)
y = df['Class']

# apply ADASYN to the dataset
adasyn = ADASYN(random_state=42)
X_adasyn, y_adasyn = adasyn.fit_resample(X, y)
df_adasyn = pd.concat([pd.DataFrame(X_adasyn), pd.DataFrame(y_adasyn)], axis=1)
```

```python
df_balanced=df_adasyn
df_balanced
```

```python
#simple random sampling
import math
import random
n=math.ceil((1.96**2 * (0.25))/(0.06**2))
random_sample = df_balanced.sample(n, random_state=42)
random_sample
accuracies1 = []

models = [LogisticRegression(), RandomForestClassifier(),GaussianNB(),DecisionTreeClassifier(),KNeighborsClassifier()]

X = random_sample.drop(['Class'], axis=1)
y = random_sample['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies1.append(acc)
```

```python
#systematic sampling
import math
import random
n=math.ceil((1.96**2 * (0.25))/(0.05**2))
start = 0 
step = len(df_balanced) // n 

systematic_sample =df_balanced.iloc[start::step].head(n)
systematic_sample
accuracies2 = []

models = [LogisticRegression(), RandomForestClassifier(),GaussianNB(),DecisionTreeClassifier(),KNeighborsClassifier()]

X = systematic_sample.drop(['Class'], axis=1)
y = systematic_sample['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies2.append(acc)
```

```python
#stratified sampling
import math
import random
n=math.ceil((1.96**2 * (0.25))/((0.1/2)**2))
stra_sampling=df_balanced.groupby('Class', group_keys=False).apply(lambda x: x.sample(n))
stra_sampling
accuracies3 = []

models = [LogisticRegression(), RandomForestClassifier(),GaussianNB(),DecisionTreeClassifier(),KNeighborsClassifier()]

X = stra_sampling.drop(['Class'], axis=1)
y = stra_sampling['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies3.append(acc)
```

```python
#cluster sampling
from imblearn.over_sampling._smote import cluster

k = int(len(df_balanced)/25)

cluster_labels = np.random.randint(0, k, size=len(df_balanced))

selected_clusters = np.random.choice(range(k), size=25, replace=False)

cluster_data = df_balanced[np.isin(cluster_labels, selected_clusters)].reset_index(drop=True)
#cluster_data
len(cluster_data)
cluster_data
accuracies4 = []

models = [LogisticRegression(), RandomForestClassifier(),GaussianNB(),DecisionTreeClassifier(),KNeighborsClassifier()]

X = cluster_data.drop(['Class'], axis=1)
y = cluster_data['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies4.append(acc)

```

```python
#snowball sampling
df_snowball = df_balanced[df_balanced['Class'] == 1].sample(n=200, random_state=42)
for i in range(6):
    referrals = df_balanced[df_balanced['Class'] == 0].sample(n=30, random_state=42)
    df_snowball = pd.concat([df_snowball, referrals])
    df_snowball = df_snowball.drop_duplicates()

df_snowball
accuracies5 = []

models = [LogisticRegression(), RandomForestClassifier(),GaussianNB(),DecisionTreeClassifier(),KNeighborsClassifier()]

X = df_snowball.drop(['Class'], axis=1)
y = df_snowball['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies5.append(acc)
```

```python
mm=['LogisticRegression','RandomForest','GaussianNB','DecisionTreeClassifier','KNN']
Accuracy_Scores=pd.DataFrame(mm)
Accuracy_Scores['SimpleRandom']=accuracies1
Accuracy_Scores['Systematic']=accuracies2
Accuracy_Scores['Stratified']=accuracies3
Accuracy_Scores['Cluster']=accuracies4
Accuracy_Scores['Snowball']=accuracies5
Accuracy_Scores
```
